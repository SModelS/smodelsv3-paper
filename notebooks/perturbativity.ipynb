```python
lambda1 = (1/4)*((mh**2 + ms**2 + (mh**2 - ms**2)*sp.cos(2*theta))/v**2)
lambda2 = (gchi**2*(mh**2 + ms**2 + (-mh**2 + ms**2)*sp.cos(2*theta)))/mZp**2
lambda3 = (gchi*(mh**2 - ms**2)*sp.sin(2*theta))/(mZp*v)
lambda1
```

```python
import numpy as np
import matplotlib.pyplot as plt

mh = 125
v = 256
gchi = 0.5  
mZp = 3000   

thetaValues = np.linspace(-np.pi/4, np.pi/4, 1000)

ms_invalid_values1 = []
ms_invalid_values2 = []
ms_invalid_values3 = []

for theta in thetaValues:
    try:
        value1 = np.sqrt((-1.0 * mh**2 * np.cos(theta)**2 + 8*np.pi*v**2) / np.sin(theta)**2)
        value2 = np.sqrt((-gchi**2 * mh**2 * np.sin(theta)**2 + 2*np.pi*mZp**2) / (gchi**2 * np.cos(theta)**2))
        if (mh**2 - 4*np.pi*mZp*v / (gchi * np.sin(2*theta))) >= 0:
           value3 = np.sqrt(mh**2 - 4*np.pi*mZp*v / (gchi * np.sin(2*theta)))
        else:
           value3 = np.nan
    except ZeroDivisionError:  # if sin(theta) is 0
        value1 = np.nan
        value2 = np.nan
        value3 = np.nan

    ms_invalid_values1.append(value1)
    ms_invalid_values2.append(value2)
    ms_invalid_values3.append(value3)

ms_invalid_values1 = np.array(ms_invalid_values1)
ms_invalid_values2 = np.array(ms_invalid_values2)
ms_invalid_values3 = np.array(ms_invalid_values3)

# Plotting
plt.figure(figsize=(10, 6))

upper_limit = 1.5 * max(np.nanmax(ms_invalid_values1), np.nanmax(ms_invalid_values2), np.nanmax(ms_invalid_values3))
plt.fill_between(thetaValues, ms_invalid_values1, upper_limit, where=(ms_invalid_values1 <= upper_limit), color='red', alpha=0.3, label=r"from $\lambda_1$")
plt.fill_between(thetaValues, ms_invalid_values2, upper_limit, where=(ms_invalid_values2 <= upper_limit), color='blue', alpha=0.3, label=r"from $\lambda_2$")
plt.fill_between(thetaValues, 0, ms_invalid_values3, where=(ms_invalid_values3 <= upper_limit), color='green', alpha=0.3, label=r"from $\lambda_3$")
plt.xlabel('Theta')
plt.ylabel('$m_s$')
#plt.title('no allowed regions')
plt.xlim(0, np.pi/4)
plt.ylim(0, 20000)
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact

mh = 125
v = 256

def plot_graph(mZp=3000, gchi=0.5):
    thetaValues = np.linspace(-np.pi/4, np.pi/4, 1000)

    ms_values1, ms_values2, ms_values3 = [], [], []

    for theta in thetaValues:
        try:
            value1 = np.sqrt((-1.0 * mh**2 * np.cos(theta)**2 + 8*np.pi*v**2) / np.sin(theta)**2)
            value2 = np.sqrt((-gchi**2 * mh**2 * np.sin(theta)**2 + 2*np.pi*mZp**2) / (gchi**2 * np.cos(theta)**2))
            if (mh**2 - 4*np.pi*mZp*v / (gchi * np.sin(2*theta))) >= 0:
                value3 = np.sqrt(mh**2 - 4*np.pi*mZp*v / (gchi * np.sin(2*theta)))
            else:
                value3 = np.nan
        except ZeroDivisionError:
            value1, value2, value3 = np.nan, np.nan, np.nan

        ms_values1.append(value1)
        ms_values2.append(value2)
        ms_values3.append(value3)

    # Plotting
    plt.figure(figsize=(10, 6))

    upper_limit = 1.5 * max(np.nanmax(ms_values1), np.nanmax(ms_values2), np.nanmax(ms_values3))
    plt.fill_between(thetaValues, ms_values1, upper_limit, where=(ms_values1 <= upper_limit), color='red', alpha=0.3, label=r"from $\lambda_1$")
    plt.fill_between(thetaValues, ms_values2, upper_limit, where=(ms_values2 <= upper_limit), color='blue', alpha=0.3, label=r"from $\lambda_2$")
    plt.fill_between(thetaValues, 0, ms_values3, where=(ms_values3 <= upper_limit), color='green', alpha=0.3, label=r"from $\lambda_3$")
    
    plt.xlabel('Theta')
    plt.ylabel('$m_s$')
    plt.xlim(0, np.pi/4)
    plt.ylim(0, 20000)
    plt.grid(True)
    plt.legend()
    plt.show()

# Interactive sliders for MZp and gchi
interact(plot_graph, mZp=(1000, 5000, 100), gchi=(0.1, 2.0, 0.1));
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define x values
x = np.linspace(0, 10, 1000)

# Calculate y = 2x
y = 2 * x

# Plotting
plt.figure(figsize=(10, 6))

# Fill the non-allowed region
plt.fill_between(x, y, 0, color='red', alpha=0.3, label='Not Allowed Region')

# Plot y = 2x
plt.plot(x, y, '-b', label='y=2x')

# Annotate and display the plot
plt.title('Non-Allowed Region for y > 2x')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
